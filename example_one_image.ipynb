```python
import numpy as np
from PIL import Image
from translateTexture import translate_texture
import matplotlib.pyplot as plt
%matplotlib inline
```

#### Load image

```python
image = np.array(Image.open('assets/Lena.jpg'))

plt.imshow(image)
plt.axis('off')
```

#### Create texture translation object 

```python
TT = translate_texture(random_crop=True, random_flip=True)
```

#### Translate to printed

```python
TT.set_source('printed')
```

```python
im_printed = TT.texture_image(image)

plt.imshow(im_printed)
plt.axis('off')
```

#### Translate to screen

```python
TT.set_source('screen')
```

```python
im_screen = TT.texture_image(image)

plt.imshow(im_screen)
plt.axis('off')
```

#### Compare the results

```python
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(15,5))
ax1.imshow(image)
ax1.set_title('digital')
ax1.axis('off')
ax2.imshow(im_printed)
ax2.set_title('printed')
ax2.axis('off')
ax3.imshow(im_screen)
ax3.set_title('screen')
ax3.axis('off')
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(15,5))
ax1.imshow(image[192:320, 192:320])
ax1.set_title('digital')
ax1.axis('off')
ax2.imshow(im_printed[192:320, 192:320])
ax2.set_title('printed')
ax2.axis('off')
ax3.imshow(im_screen[192:320, 192:320])
ax3.set_title('screen')
ax3.axis('off')
```
